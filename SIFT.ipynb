```python
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

# Reading image

```python
img1 = cv2.imread('constrasted_image1.png')
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

# Keypoints

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
```

# Print

```python
img_1 = cv2.drawKeypoints(gray1,keypoints_1,img1)
plt.imshow(img_1)
```

# Deuxieme test

```python
img = cv2.imread('constrasted_image2.png')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp = sift.detect(gray,None)
```

```python
img=cv2.drawKeypoints(gray,kp,img,color=(0,0,255))
```

```python
cv2.imwrite('sift_keypoints_contrasted_im_2.jpg',img)
```

```python
img=cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS,color=(0,0,255))
cv2.imwrite('sift_keypoints_gradient_contrasted_im_2.jpg',img)
```

```python
new_im = cv2.imread('sift_keypoints_gradient.jpg')
plt.imshow(new_im)
```
